## Plotting polar graphs using `plot` function

We can graph the polar function $r = f(\theta)$ using `plot` function upon parametrization.

$$ 
x = f(\theta) \cos(\theta)
\quad\text{and}\quad
y = f(\theta) \sin(\theta).
$$

For instance, a cardioid (a heart-shaped figure) $r = 1 + \cos(\theta + \pi/2)$ for $\theta \in [0, 2\pi]$ can be plotted as follows:

```matlab
f = @(theta) 1 + cos(theta + pi/2);
theta = linspace(0, 2*pi, 361);
x = f(theta) .* cos(theta);
y = f(theta) .* sin(theta);
plot(x, y)
axis image
```

## Stirling's formula

The following formula approximates $n!$ for large values of $n$:

$$ n ! \approx \sqrt{2\pi n} \left( \frac{n}{e} \right)^n. $$

We found out from trial and error that MATLAB can handle $n!$ for $n$ up to 170. We plotted the absolute and relative errors between `factorial` routine and the Stirling's approximation formula as follows:

```matlab
%% anonymous function for Stirling's formula
stirn = @(n) sqrt(2*pi*n) .* (n/exp(1)) .^ n;

%% vectorization of data
n_vec = [1:170]; 
err_abs = abs( stirn(n) - factorial(n) );
err_rel = abs( stirn(n)./factorial(n) - 1 );

%% plot using subplot
% note the use of "semilogy" function below to scale the vertical axes. 

subplot(121); % same as "subplot(1,2,2)"
semilogy(n, err_abs);
xlabel('n')
title('absolute error')

subplot(122); % same as "subplot(1,2,1)"
semilogy(n, err_rel);
xlabel('n')
title('relative error')
```
